```python
# #### Extracting home pages tasg

tags_from_website= []
df_website_non_vc['tags_in_home'] = 0
pattern = re.compile(r'(,\s){2,}')
for web_link in range(0,len(df_website_non_vc)):
    tags = []

#     web_name = "http://www.crescent.com"
    url = "http://"+str(df_website_non_vc['url'][web_link]) # Url of V
    print(url)
    hdr = {'User-Agent': 'Mozilla/7.0'}
    req = Request(url,headers=hdr)
    a_tag = str()
    try:
        page = urlopen(req)
        soup = BeautifulSoup(page)
        tags = soup.find_all('a') ### Finding a  tags 
        ## getting text
        a_tag = BeautifulSoup(str(tags)).get_text()
        a_tag = a_tag.replace('\n', ' ').replace('\r', '').replace('[', ' ').replace(']', ' ').replace('\t', ' ')
        a_tag = ' '.join(a_tag.split())
        a_tag = ''.join(a_tag)
        a_tag = re.sub(' +', ' ', a_tag)
        a_tag = re.sub(', ,', ', ', a_tag)
        a_tag = re.sub(' , ', ',', a_tag)
        a_tag = re.sub(' ,', ',', a_tag)
        a_tag = re.sub(', ', ',', a_tag)
        a_tag = re.sub(pattern, ', ', a_tag)
        a_tag = a_tag[1:]
        a_tag = a_tag[0:-1]
        print("do")
        print(str(a_tag))
    except:
        print("Not Found")

    
    df_website_non_vc['tags_in_home'][web_link] = str(a_tag)
    
            
#     site.append(a_tag)
```

### Extracting hyperlinks Basically h_tags from every home page possible:

#### Loop for Href link finder :
Multiple web links are parsed in a loop to extract html tags where we search all a_tags & h_tags inside it.

```python
# from urllib.request import Request, urlopen

df_website_non_vc['hrefs'] = ''
for url in range(0,int(len(df_website_non_vc))):
    URL = str("http://"+str(df_website_non_vc.iloc[url]['url']))
    print(URL)
    try: # Url of V

        hdr = {'User-Agent': 'Mozilla/7.0'}
        req = Request(URL,headers=hdr)
        html_page = urlopen(req)
        
        soup = BeautifulSoup(html_page)
        print("website accecible")

        links = []

        for link in soup.findAll('a'):
            links_found = []
            try:
                a_tag = link.get('href')
                if a_tag.find('https://')==0:
                    print("verified url")
                    links_found.append(a_tag)
                    links.append(link.get('href'))

            except:
                print("Not Found")
                continue
            all_links_list = []   
            for link_check in links_found:
                if link_check!=[]:
                    all_links_list.append(links_found)
                else:
                    print("empty a tag")
            all_links_list.append(all_links_list)
            
        df_website_non_vc['hrefs'][url] = str(links)
    except:
        print("Not a Valid address")
```
