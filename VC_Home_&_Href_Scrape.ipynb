### Method 2:

```python
from bs4 import BeautifulSoup
from bs4.element import Comment
from urllib.request import Request, urlopen
import re


def tag_visible(element):
    if element.parent.name in ['style', 'script', 'head', 'title', 'meta', '[document]']:
        return False
    if isinstance(element, Comment):
        return False
    return True


def text_from_html(body):
    soup = BeautifulSoup(body, 'html.parser')
    texts = soup.findAll(text=True)
    visible_texts = filter(tag_visible, texts)  
    return u" ".join(t.strip() for t in visible_texts)
```

```python
df_website_vc['home_page'] = ''
for url in range(0,len(df_website_vc)):
    
    URL = str("http://"+str(df_website_vc.iloc[url]['url']))
    hdr = {'User-Agent': 'Mozilla/7.0'}
    try:
        req = Request(URL,headers=hdr)
        html_page = urlopen(req)
        print("website accecible")
        a = text_from_html(html_page)
        cleand_text = ' '.join(a.split())
        df_website_vc['home_page'][url] =  str(cleand_text)
    except:
        print("Not found")
```

### Extracting hyperlinks Basically h_tags from every home page possible:

#### Loop for Href link finder :
Multiple web links are parsed in a loop to extract html tags where we search all a_tags & h_tags inside it.

```python
from urllib.request import Request, urlopen
import re

df_website_vc['hrefs'] = ''
for url in range(0,int(len(df_website_vc))):
    URL = str("http://"+str(df_website_vc.iloc[url]['url']))
    print(URL)
    try: # Url of V

        hdr = {'User-Agent': 'Mozilla/7.0'}
        req = Request(URL,headers=hdr)
        html_page = urlopen(req)
        
        soup = BeautifulSoup(html_page)
        print("website accecible")

        links = []

        for link in soup.findAll('a'):
            links_found = []
            try:
                a_tag = link.get('href')
                if a_tag.find('https://')==0:
                    print("verified url")
                    links_found.append(a_tag)
                    links.append(link.get('href'))

            except:
                print("Not Found")
                continue
            all_links_list = []   
            for link_check in links_found:
                if link_check!=[]:
                    all_links_list.append(links_found)
                else:
                    print("empty a tag")
            all_links_list.append(all_links_list)
            
        df_website_vc['hrefs'][url] = str(links)
    except:
        print("Not a Valid address")
```
